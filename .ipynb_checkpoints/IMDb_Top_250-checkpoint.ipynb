```python
# import the modules we will need to this project
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
```

```python
# Create a function that will be used later to check connection status
def check_response(resp_obj):
    # if the requests object returns 200 we have a connection
    try:
        if resp_obj.status_code == 200:
            print(f"Connection OK: {resp_obj.status_code}")
    except:
        # If we are not able to access the page what was the HTTP error code
        # 4xx: CLient Server, 5xx: Server Error will be the most common
        print(f"Error: Not 200 return value was: {resp_obj}")
```

```python
# Convert the address to a raw string 
url_top_250 = r"https://www.imdb.com/chart/top/"

# Using requests get address 
page_250 = requests.get(url_top_250)

# call our function from above and print the connection status
check_response(page_250)
```

```python
# Using bs4 we can create the Soup
soup = BeautifulSoup(page_250.text, 'html.parser')
type(soup)
```

```python
# soup.body will access all of the HTML within the body tags <body>...</body>
# The table is contained there so no reason to return all of the information within the header, too
body_tag = soup.body
# confirm the element type
type(body_tag)
```

```python
# Here we use find_all looking for table element with class  and attribute class name 'chart full-width'
table = body_tag.find_all('table',{'class': 'chart full-width'})
print(type(table))
# Check the length of variable tables it should contain 4 tables
print(f"How many tables???? {len(table)}")
```

```python
#to see table data remove # from table
movies = table[0]
type(movies)
print(movies)
```

```python
# Find all 250 <td> with class of "titleColumn" This will hold title in "a" tag and year in span
movie_details = movies.find_all('td')
movie_details # this is a list
```

```python
print(f'What is the lenght of title_year {len(movie_details)}')
```

```python
title = movie_details[1].find('a', href=True, title=True)
title
```

```python
rate = movie_details[7].find('strong', title=True)
rate.text
```

```python
movie_titles = []
movie_years = []
IMDb_ratings = []

for i in range(len(movie_details)):
    title = movie_details[i].find('a', href=True, title=True)
    year = movie_details[i].find('span', {'class' : 'secondaryInfo'})
    if movie_details[i].find('strong', title=True) != None:
        IMDb_ratings.append(movie_details[i].find('strong', title=True).text)
#     rating = movie_details[i].find('strong', title=True)
    try:
        movie_titles.append(title.text)
        movie_years.append(year.text)

    except:
        # if None pass it will throw an error
        pass

# should be 250 if this works 
print(len(movie_titles))
print(len(movie_years))
print(len(IMDb_ratings))
print(len(IMDb_ratings))
```
